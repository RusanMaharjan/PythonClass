```python
import csv
csv_columns = ['Name', 'Age', 'Address']
dict = [
    {'Name':'Abc', 'Age':23,'Address':'Kathmandu'},
    {'Name':'Def', 'Age':20,'Address':'Lalitpur'}
]
try:
    with open("data.csv", 'w') as csvfile:
        writer = csv.DictWriter(csvfile, fieldnames = csv_columns)
        writer.writeheader()
        for data in dict:
            writer.writerow(data)
except IOError:
    print("I/O error")
```
